## 1.Histogram for a normal distribution.

Prompt: Python code for representation of a normal distribution in form of a histogram

```
import numpy as np
import matplotlib.pyplot as plt

# Take input from the user
try:
    mu = float(input("Enter the mean: "))
    sigma = float(input("Enter the standard deviation: "))
    n_samples = int(input("Enter the number of samples: "))

    # Generate samples
    data = np.random.normal(loc=mu, scale=sigma, size=n_samples)

    # Create the plot
    plt.figure(figsize=(8, 5))
    plt.hist(data, bins=30, color='lightcoral', edgecolor='black', alpha=0.7)
    plt.title(f'Histogram of a Normal Distribution (mean={mu}, std={sigma})')
    plt.xlabel('Value')
    plt.ylabel('Frequency')
    plt.grid(axis='y', alpha=0.3)
    plt.show()
except ValueError:
    print("Invalid input. Please enter numerical values.")
```

## 2.Computation of mean, median, variance, and standard deviation for a dataset.

Prompt: Python code computation of mean, medain, variance and standard deviation for the data provided by user.


```
import numpy as np

# Take input from the user
try:
    user_input = input("Enter numbers separated by spaces or commas: ")
    # Handle both space and comma separation
    data_list = [float(x) for x in user_input.replace(',', ' ').split()]
    data = np.array(data_list)

    if len(data) > 0:
        # Compute statistics
        mean_val = np.mean(data)
        median_val = np.median(data)
        variance_val = np.var(data)
        std_dev_val = np.std(data)

        # Display results
        print(f"Dataset: {data}")
        print(f"Mean: {mean_val:.4f}")
        print(f"Median: {median_val:.4f}")
        print(f"Variance: {variance_val:.4f}")
        print(f"Standard Deviation: {std_dev_val:.4f}")
    else:
        print("Dataset is empty. Please enter some numbers.")
except ValueError:
    print("Invalid input. Please enter numerical values separated by spaces or commas.")
```

3.Simulation of a coin toss experiment and computing its probabilities.

Prompt: Python program for a coin toss experiment and compute its probability>

```
import random

try:
    n_tosses = int(input("Enter the number of tosses: "))

    if n_tosses > 0:
        # Simulate tosses (0 for Heads, 1 for Tails)
        results = [random.randint(0, 1) for _ in range(n_tosses)]

        heads_count = results.count(0)
        tails_count = results.count(1)

        prob_heads = heads_count / n_tosses
        prob_tails = tails_count / n_tosses

        print(f"Total Tosses: {n_tosses}")
        print(f"Heads: {heads_count} (Probability: {prob_heads:.4f})")
        print(f"Tails: {tails_count} (Probability: {prob_tails:.4f})")
    else:
        print("Please enter a positive integer.")
except ValueError:
    print("Invalid input. Please enter a numerical integer.")
```
