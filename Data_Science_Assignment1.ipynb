# 1.Factorial of number.


Prompt: Python program for computation of factorial number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage with user input:
try:
    user_input = input("Enter a non-negative integer to compute its factorial: ")
    number = int(user_input)
    print(f"The factorial of {number} is: {factorial(number)}")
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# number = 5
# print(f"The factorial of {number} is: {factorial(number)}")

# number = 0
# print(f"The factorial of {number} is: {factorial(number)}")

# number = -3
# print(f"The factorial of {number} is: {factorial(number)}")
```

# 2.Display table for given number


Prompt: Python program for displaying multiplication table for given number, take input from user

```python
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage with user input:
try:
    user_input = input("Enter an integer to display its multiplication table: ")
    num = int(user_input)
    display_multiplication_table(num)
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# num = 7
# display_multiplication_table(num)

# print("\n") # Add a newline for better readability between tables

# num_two = 12
# display_multiplication_table(num_two, limit=5) # Display table up to 5 for number 12
```

# 3.Identifying  the given number is odd or even

Prompt: Python code for identifying if the given number is odd or even, take input from user

```python
def check_odd_even(number):
    if number % 2 == 0:
        return f"{number} is an even number."
    else:
        return f"{number} is an odd number."

# Example usage with user input:
try:
    user_input = input("Enter an integer to check if it's odd or even: ")
    number = int(user_input)
    print(check_odd_even(number))
except ValueError:
    print("Invalid input. Please enter an integer.")

# Optional: You can still include the previous examples for testing if you like:
# print(check_odd_even(4))
# print(check_odd_even(7))
# print(check_odd_even(0))
# print(check_odd_even(-2))
# print(check_odd_even(-5))
```

# 4.Identifying the entered number is palindrome or not.

Prompt: Python code for identifying if the given number is palindrome or not, take input from user

```python
def is_palindrome_number(number):
    # Convert the number to a string to easily reverse and compare
    num_str = str(number)
    # Compare the string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage with user input:
try:
    user_input = input("Enter an integer to check if it's a palindrome: ")
    number = int(user_input)
    print(is_palindrome_number(number))
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# print(is_palindrome_number(121))
# print(is_palindrome_number(123))
# print(is_palindrome_number(12321))
# print(is_palindrome_number(10))
# print(is_palindrome_number(-121)) # Palindrome check usually applies to non-negative numbers, but this handles string reversal
```

## 5.Developing  an Expert System that provides simple decision-making.

Prompt: Python code for developing an expert system which provide simple decision-making for course selection, take input from user

```python
def run_expert_system_course(facts, rules):
    print("\n--- Course Selection Expert System Execution ---")
    print(f"Initial Facts: {facts}")

    new_facts_added = True
    iteration = 0
    course_suggestions = []

    while new_facts_added:
        new_facts_added = False
        iteration += 1
        for rule_name, rule_body in rules.items():
            conditions = rule_body['conditions']
            suggestion = rule_body['suggestion']

            all_conditions_met = True
            for condition in conditions:
                if condition not in facts:
                    all_conditions_met = False
                    break

            if all_conditions_met and suggestion not in course_suggestions:
                course_suggestions.append(suggestion)
                new_facts_added = True
                # For simplicity, we just add the suggestion directly to facts to avoid re-triggering the same rule
                # In a more complex system, 'suggestion' might be a new 'fact' that triggers other rules.
                # For this simple case, we directly collect the suggestions.
                print(f"  Rule '{rule_name}' fired. Suggested course: '{suggestion}'")

    print(f"\nFinal Suggestions after {iteration} iterations: {course_suggestions}")
    print("--- End of Course Selection Expert System Execution ---")
    return course_suggestions

# Define the Knowledge Base (Rules for course suggestions)
rules_course_selection = {
    "rule_math_interest_data_science": {
        "conditions": ["interest_math", "career_data_science"],
        "suggestion": "Suggested_Course: Introduction to Data Science"
    },
    "rule_programming_interest_web_dev": {
        "conditions": ["interest_programming", "career_web_development"],
        "suggestion": "Suggested_Course: Web Development Fundamentals"
    },
    "rule_art_interest_graphic_design": {
        "conditions": ["interest_art", "career_graphic_design"],
        "suggestion": "Suggested_Course: Graphic Design Basics"
    },
    "rule_biology_interest_medicine": {
        "conditions": ["interest_biology", "career_medicine"],
        "suggestion": "Suggested_Course: Human Anatomy and Physiology"
    },
    "rule_history_interest_journalism": {
        "conditions": ["interest_history", "career_journalism"],
        "suggestion": "Suggested_Course: Media and Communication Studies"
    },
    "rule_creative_writing_literature": {
        "conditions": ["interest_writing", "interest_literature"],
        "suggestion": "Suggested_Course: Creative Writing Workshop"
    },
    "rule_problem_solving_engineering": {
        "conditions": ["interest_problem_solving", "career_engineering"],
        "suggestion": "Suggested_Course: Engineering Principles"
    },
    "rule_default_programming": {
        "conditions": ["interest_programming"],
        "suggestion": "Suggested_Course: Python Programming Basics"
    },
    "rule_default_humanities": {
        "conditions": ["interest_general_humanities"],
        "suggestion": "Suggested_Course: Introduction to Philosophy"
    }
}

# Example Usage with user input:
try:
    print("\nHello! I can help you select a course.")
    user_input_interests = input("What are your main interests? (e.g., interest_math, interest_programming, interest_art, interest_biology, interest_history, interest_writing, interest_literature, interest_problem_solving, interest_general_humanities - separate with commas): ")
    user_input_career = input("What is your career goal? (e.g., career_data_science, career_web_development, career_graphic_design, career_medicine, career_journalism - leave blank if none): ")

    # Clean and convert user input into a list of facts
    initial_facts_for_courses = []
    if user_input_interests.strip():
        initial_facts_for_courses.extend([fact.strip().lower() for fact in user_input_interests.split(',') if fact.strip()])
    if user_input_career.strip():
        initial_facts_for_courses.append(user_input_career.strip().lower())

    if not initial_facts_for_courses:
        print("No valid preferences entered. I'll suggest a general course.")
        initial_facts_for_courses.append("interest_general_humanities") # Default if no input

    print("\n*** Running Course Selection Expert System ***")
    suggestions = run_expert_system_course(initial_facts_for_courses.copy(), rules_course_selection)

    if suggestions:
        print("\nBased on your preferences, I suggest the following courses:")
        for suggestion in suggestions:
            print(f"- {suggestion.replace('Suggested_Course: ', '')}")
    else:
        print("\nI couldn't find a specific course suggestion based on your input. Try providing more specific interests or career goals.")

except Exception as e:
    print(f"An error occurred: {e}")
```
