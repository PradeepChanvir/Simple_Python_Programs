# 1.Factorial of number.


Prompt: Python program for computation of factorial number

```python
def factorial(n):
    if n < 0:
        return "Factorial is not defined for negative numbers"
    elif n == 0:
        return 1
    else:
        result = 1
        for i in range(1, n + 1):
            result *= i
        return result

# Example usage with user input:
try:
    user_input = input("Enter a non-negative integer to compute its factorial: ")
    number = int(user_input)
    print(f"The factorial of {number} is: {factorial(number)}")
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# number = 5
# print(f"The factorial of {number} is: {factorial(number)}")

# number = 0
# print(f"The factorial of {number} is: {factorial(number)}")

# number = -3
# print(f"The factorial of {number} is: {factorial(number)}")
```

# 2.Display table for given number


Prompt: Python program for displaying multiplication table for given number, take input from user

```python
def display_multiplication_table(number, limit=10):
    print(f"Multiplication Table for {number}:")
    for i in range(1, limit + 1):
        print(f"{number} x {i} = {number * i}")

# Example usage with user input:
try:
    user_input = input("Enter an integer to display its multiplication table: ")
    num = int(user_input)
    display_multiplication_table(num)
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# num = 7
# display_multiplication_table(num)

# print("\n") # Add a newline for better readability between tables

# num_two = 12
# display_multiplication_table(num_two, limit=5) # Display table up to 5 for number 12
```

# 3.Identifying  the given number is odd or even

Prompt: Python code for identifying if the given number is odd or even, take input from user

```python
def check_odd_even_bitwise(number):
    # A number is even if its least significant bit is 0
    # A number is odd if its least significant bit is 1
    if (number & 1) == 0:
        return f"{number} is an even number (using bitwise check)."
    else:
        return f"{number} is an odd number (using bitwise check)."

# Example usage with user input:
try:
    user_input = input("Enter an integer to check if it's odd or even: ")
    number = int(user_input)
    print(check_odd_even_bitwise(number))
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# print(check_odd_even_bitwise(4))
# print(check_odd_even_bitwise(7))
# print(check_odd_even_bitwise(0))
# print(check_odd_even_bitwise(-2))
# print(check_odd_even_bitwise(-5))
```

# 4.Identifying the entered number is palindrome or not.

Prompt: Python code for identifying if the given number is palindrome or not, take input from user

```python
def is_palindrome_number(number):
    # Convert the number to a string to easily reverse and compare
    num_str = str(number)
    # Compare the string with its reversed version
    if num_str == num_str[::-1]:
        return f"{number} is a palindrome."
    else:
        return f"{number} is not a palindrome."

# Example usage with user input:
try:
    user_input = input("Enter an integer to check if it's a palindrome: ")
    number = int(user_input)
    print(is_palindrome_number(number))
except ValueError:
    print("Invalid input. Please enter an integer.")

# You can still include the previous examples for testing if you like:
# print(is_palindrome_number(121))
# print(is_palindrome_number(123))
# print(is_palindrome_number(12321))
# print(is_palindrome_number(10))
# print(is_palindrome_number(-121)) # Palindrome check usually applies to non-negative numbers, but this handles string reversal
```

## 5.Developing  an Expert System that provides simple decision-making.

Prompt: Python code for developing an expert system which provide simple decision-making , take input from user

```python
def run_expert_system(facts, rules):
    print("\n--- Expert System Execution ---")
    print(f"Initial Facts: {facts}")
    print(f"Rules: {rules}")

    # Keep track of newly deduced facts in each pass
    new_facts_added = True
    iteration = 0

    while new_facts_added:
        new_facts_added = False
        iteration += 1
        print(f"\nIteration {iteration}:")
        for rule_name, rule_body in rules.items():
            conditions = rule_body['conditions']
            conclusion = rule_body['conclusion']

            # Check if all conditions for the rule are met
            all_conditions_met = True
            for condition in conditions:
                if condition not in facts:
                    all_conditions_met = False
                    break

            # If conditions are met and conclusion is not already a fact, add it
            if all_conditions_met and conclusion not in facts:
                facts.append(conclusion)
                new_facts_added = True
                print(f"  Rule '{rule_name}' fired. Added fact: '{conclusion}'")

    print(f"\nFinal Facts after {iteration} iterations: {facts}")
    print("--- End of Expert System Execution ---")
    return facts

# Define the Knowledge Base (Rules remain static for this simple example)
rules_1 = {
    "rule_1": {
        "conditions": ["is_sunny", "temperature_warm"],
        "conclusion": "go_outside"
    },
    "rule_2": {
        "conditions": ["go_outside"],
        "conclusion": "wear_sunglasses"
    },
    "rule_3": {
        "conditions": ["temperature_warm"],
        "conclusion": "drink_water"
    },
    "rule_4": {
        "conditions": ["is_raining"],
        "conclusion": "stay_indoors"
    },
    "rule_5": {
        "conditions": ["stay_indoors"],
        "conclusion": "read_book"
    }
}

# Example Usage with user input:
try:
    user_input = input("Enter initial facts, separated by commas (e.g., is_sunny,temperature_warm): ")
    # Clean and convert user input into a list of facts
    initial_facts_from_user = [fact.strip() for fact in user_input.split(',') if fact.strip()]

    if not initial_facts_from_user:
        print("No valid facts entered. Running expert system with an empty fact base.")

    print("\n*** Running Expert System with User-Provided Facts ***")
    final_facts_user = run_expert_system(initial_facts_from_user.copy(), rules_1)

except Exception as e:
    print(f"An error occurred: {e}")

# You can still include the previous examples for testing if you like:
# print("\n*** Running Example 1 (Sunny and Warm) ***")
# initial_facts_1 = [
#     "is_sunny",
#     "temperature_warm"
# ]
# final_facts_1 = run_expert_system(initial_facts_1.copy(), rules_1)

# print("\n*** Running Example 2 (Raining and Cold) ***")
# initial_facts_2 = [
#     "is_raining",
#     "temperature_cold"
# ]
# final_facts_2 = run_expert_system(initial_facts_2.copy(), rules_1)

# print("\n*** Running Example 3 (Cloudy) ***")
# initial_facts_3 = [
#     "is_cloudy"
# ]
# final_facts_3 = run_expert_system(initial_facts_3.copy(), rules_1)
```
